```python

1. What is a lambda function in Python, and how does it differ from a regular function?

A lambda function in Python is a small anonymous function. It is defined using the following syntax:

Python
lambda arguments: expression

The arguments part of the lambda function definition specifies the arguments that the function can accept. The expression part of the lambda function definition specifies the code that the function will execute.

Lambda functions are different from regular functions in that they are anonymous and can only contain one expression. Lambda functions are often used in conjunction with other functions, such as the map(), filter(), and reduce() functions.

2. Can a lambda function in Python have multiple arguments? If yes, how can you define and use
them?

Yes, a lambda function in Python can have multiple arguments. To define a lambda function with multiple arguments, you simply list the arguments in a comma-separated list before the expression. For example, the following lambda function takes two numbers as arguments and returns the sum of the two numbers:

Python
sum_function = lambda a, b: a + b

To use a lambda function with multiple arguments, you simply pass the arguments to the function like you would with any other Python function. For example, the following code shows how to use the sum_function lambda function:

Python
sum = sum_function(1, 2)

print(sum)

Output:

3
3. How are lambda functions typically used in Python? Provide an example use case.

Lambda functions are typically used to perform simple tasks that do not require a full-fledged function. For example, you could use a lambda function to sort a list of numbers, filter a list of items, or perform a mathematical operation on a list of values.

Here is an example of how to use a lambda function to sort a list of numbers:

Python
numbers = [1, 5, 3, 2, 4]

sorted_numbers = sorted(numbers, key=lambda x: x)

print(sorted_numbers)

Output:

[1, 2, 3, 4, 5]
4. What are the advantages and limitations of lambda functions compared to regular functions in
Python?

Advantages of lambda functions:

They are concise and easy to read.
They are efficient, as they are only evaluated when they are called.
They can be used inline, without having to define a named function first.
Limitations of lambda functions:

They can only contain one expression.
They cannot be used to define recursive functions.
They can be difficult to debug, as they do not have a name.
5. Are lambda functions in Python able to access variables defined outside of their own scope?
Explain with an example.

Yes, lambda functions in Python are able to access variables defined outside of their own scope. This is because lambda functions are created in the same scope as the function that they are called in.

Here is an example of a lambda function that accesses a variable defined outside of its own scope:

Python
global_variable = 10

def my_function():
  lambda_function = lambda: global_variable
  return lambda_function

lambda_function = my_function()
result = lambda_function()

print(result)

Output:

10
6. Write a lambda function to calculate the square of a given number.

Python
square_function = lambda x: x ** 2

This lambda function takes a number as an argument and returns the square of the number.

7. Create a lambda function to find the maximum value in a list of integers.

Python
max_function = lambda numbers: max(numbers)

This lambda function takes a list of integers as an argument and returns the maximum value in the list.

8. Implement a lambda function to filter out all the even numbers from a list of integers.

Python
filter_even_numbers = lambda numbers: [number for number in numbers if number % 2 != 0]

This lambda function takes a list of integers as an argument and returns a new list containing all of the odd numbers from the original list.

9. Write a lambda function to sort a list of strings in ascending order based on the length of each
string.

Python
sort_strings_by_length = lambda strings: sorted(strings, key=len)

This lambda function takes a list of strings as an argument and returns a new list containing the strings sorted in ascending order based on their length.


10. Create a lambda function that takes two lists as input and returns a new list containing the
common elements between the two lists.

Python
find_common_elements = lambda list1, list2: [element for element in list1 if element in list2]

This lambda function takes two lists as arguments and returns a new list containing all of the elements that are common to both lists.

Here is an example of how to use the find_common_elements lambda function:

Python
list1 = [1, 2, 3, 4, 5]
list2 = [3, 4, 5, 6, 7]

common_elements = find_common_elements(list1, list2)

print(common_elements)

Output:

[3, 4, 5]
11. Write a recursive function to calculate the factorial of a given positive integer.

Python
def factorial(n):
  if n == 0:
    return 1
  else:
    return n * factorial(n - 1)

This recursive function takes a positive integer as an argument and returns the factorial of that integer.

Here is an example of how to use the factorial function:

Python
n = 5

factorial_of_n = factorial(n)

print(factorial_of_n)

Output:

120
12. Implement a recursive function to compute the nth Fibonacci number.

Python
def fibonacci(n):
  if n == 0 or n == 1:
    return n
  else:
    return fibonacci(n - 1) + fibonacci(n - 2)

This recursive function takes an integer as an argument and returns the nth Fibonacci number.

Here is an example of how to use the fibonacci function:

Python
n = 10

fibonacci_number = fibonacci(n)

print(fibonacci_number)

Output:

55
13. Create a recursive function to find the sum of all the elements in a given list.

Python
def sum_list(list1):
  if len(list1) == 0:
    return 0
  else:
    return list1[0] + sum_list(list1[1:])

This recursive function takes a list as an argument and returns the sum of all of the elements in the list.

Here is an example of how to use the sum_list function:

Python
list1 = [1, 2, 3, 4, 5]

sum_of_list = sum_list(list1)

print(sum_of_list)

Output:

15
14. Write a recursive function to determine whether a given string is a palindrome.

Python
def is_palindrome(string):
  if len(string) <= 1:
    return True
  else:
    return string[0] == string[-1] and is_palindrome(string[1:-1])

This recursive function takes a string as an argument and returns True if the string is a palindrome, False otherwise.

Here is an example of how to use the is_palindrome function:

Python
string = "madam"

is_palindrome_string = is_palindrome(string)

print(is_palindrome_string)

Output:

True
15. Implement a recursive function to find the greatest common divisor (GCD) of two positive integers.

Python
def gcd(a, b):
  if b == 0:
    return a
  else:
    return gcd(b, a % b)

This recursive function takes two positive integers as arguments and returns the greatest common divisor (GCD) of the two integers.

Here is an example of how to use the gcd function:

Python
a = 10
b = 5

greatest_common_divisor = gcd(a, b)

print(greatest_common_divisor)


Output:

5
```
