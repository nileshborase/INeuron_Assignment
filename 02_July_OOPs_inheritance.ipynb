```python
1. Explain what inheritance is in object-oriented programming and why it is used.

Inheritance is a mechanism in object-oriented programming (OOP) that allows one class to inherit the attributes and methods of another class. This is useful because it allows us to reuse code and avoid duplicating code. For example, we can create a Vehicle class that defines the common attributes and methods of all vehicles, such as a name, a color, and a speed. We can then create subclasses of the Vehicle class, such as Car and Truck, that inherit the attributes and methods of the Vehicle class and add their own unique attributes and methods.

2. Discuss the concept of single inheritance and multiple inheritance, highlighting their
differences and advantages.

Single inheritance is when a class inherits from only one other class. Multiple inheritance is when a class inherits from more than one other class.

The main difference between single inheritance and multiple inheritance is that with single inheritance, there is a clear hierarchy of classes. The Vehicle class is the parent class of the Car class and the Truck class. With multiple inheritance, there is no clear hierarchy. The Car class and the Truck class could both inherit from the Vehicle class, but they could also inherit from other classes.

The main advantage of single inheritance is that it is easier to understand and maintain. The main advantage of multiple inheritance is that it allows us to reuse code from multiple classes.

3. Explain the terms "base class" and "derived class" in the context of inheritance.

The Vehicle class is called the base class, and the Car class and the Truck class are called derived classes. The derived classes inherit the attributes and methods of the base class.

4. What is the significance of the "protected" access modifier in inheritance? How does
it differ from "private" and "public" modifiers?

The protected access modifier is used to make attributes and methods accessible to the base class and its derived classes. The private access modifier is used to make attributes and methods accessible only to the class in which they are declared. The public access modifier is used to make attributes and methods accessible to all classes.

The protected access modifier is useful when we want to allow derived classes to access the attributes and methods of the base class, but we don't want other classes to be able to access them.

5. What is the purpose of the "super" keyword in inheritance? Provide an example.

The super keyword is used to refer to the base class of a derived class. It can be used to call the methods of the base class.

For example, the following code shows how the super keyword can be used to call the start() method of the Vehicle class from the Car class:

class Car(Vehicle):
    def start(self):
        super().start()
        print("The car has started.")
In this code, the start() method of the Car class calls the start() method of the Vehicle class. This ensures that the Car class inherits the functionality of the Vehicle class.
```

```python
Question No. 6
Answer-

class Vehicle:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def start(self):
        print("The vehicle has started.")

class Car(Vehicle):
    def __init__(self, make, model, year, fuel_type):
        super().__init__(make, model, year)
        self.fuel_type = fuel_type

    def drive(self):
        print("The car is driving.")
```

```python
Question No. 7
Answer-

class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

class Manager(Employee):
    def __init__(self, name, salary, department):
        super().__init__(name, salary)
        self.department = department

class Developer(Employee):
    def __init__(self, name, salary, programming_language):
        super().__init__(name, salary)
        self.programming_language = programming_language
```

```python
Question No. 8
Answer-

class Shape:
    def __init__(self, color, border_width):
        self.color = color
        self.border_width = border_width

class Rectangle(Shape):
    def __init__(self, color, border_width, length, width):
        super().__init__(color, border_width)
        self.length = length
        self.width = width

class Circle(Shape):
    def __init__(self, color, border_width, radius):
        super().__init__(color, border_width)
        self.radius = radius
```

```python
Question No. 10
Answer-

class BankAccount:
    def __init__(self, account_number, balance):
        self.account_number = account_number
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        self.balance -= amount

class SavingsAccount(BankAccount):
    def calculate_interest(self):
        interest_rate = 0.05
        interest = self.balance * interest_rate
        self.balance += interest

class CheckingAccount(BankAccount):
    def deduct_fees(self):
        fee = 5
        self.balance -= fee
```
